```python
import pandas as pd
file_path = '/content/drive/MyDrive/winequality-red.csv'
df = pd.read_csv(file_path)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# =========================================================
# 1. ðŸ“Š Bar Chart: Distribution of Wine Quality
# =========================================================
plt.figure(figsize=(8, 5))
df['quality'].value_counts().sort_index().plot(
    kind='bar',
    color='skyblue',
    edgecolor='black'
)
plt.title('Distribution of Wine Quality Ratings')
plt.xlabel('Quality Rating (Score from 3 to 8)')
plt.ylabel('Count of Wines')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# =========================================================
# 2. ðŸ“¦ Box Plot: Volatile Acidity vs. Quality
# =========================================================
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='volatile acidity', data=df, palette='viridis')
plt.title('Distribution of Volatile Acidity by Wine Quality Rating')
plt.xlabel('Quality Rating')
plt.ylabel('Volatile Acidity')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()


# =========================================================
# 3. âœ¨ Scatter Plot: Alcohol vs. Quality
# =========================================================
plt.figure(figsize=(10, 6))
# Using a small alpha (0.3) helps visualize dense areas
plt.scatter(df['quality'], df['alcohol'], alpha=0.3, color='darkred')
plt.title('Relationship Between Alcohol Content and Quality Rating')
plt.xlabel('Quality Rating (Score from 3 to 8)')
plt.ylabel('Alcohol Content (% by volume)')
# Ensure only the discrete quality scores are on the x-axis
plt.xticks(df['quality'].unique())
plt.grid(axis='both', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
