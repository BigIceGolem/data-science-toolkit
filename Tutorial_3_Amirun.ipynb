```python
## Two random Numpy arrays and dot product


# Set a fixed seed for reproducible results
np.random.seed(42)

# 1. Define the dimensions
n = 3
p = 4
m = 2

print(f"Creating two random arrays:")
print(f"Array 1 (A) will have shape n x p = {n} x {p}")
print(f"Array 2 (B) will have shape p x m = {p} x {m}")
print("-" * 30)

# 2. Create the two random numpy arrays
# Using np.random.random() for floats between 0.0 and 1.0
A = np.random.random((n, p))
B = np.random.random((p, m))

print(f"Array A (shape {A.shape}):")
print(A)
print(f"\nArray B (shape {B.shape}):")
print(B)
print("-" * 30)

# 3. Compute the dot product
# The dot product of (n, p) and (p, m) results in an (n, m) matrix
dot_product = np.dot(A, B)

# An alternative modern syntax is:
# dot_product = A @ B

print(f"Dot Product (A . B) (shape {dot_product.shape}):")
print(dot_product)
```
